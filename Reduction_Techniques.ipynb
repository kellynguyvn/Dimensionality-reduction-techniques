```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE, MDS, Isomap, LocallyLinearEmbedding, SpectralEmbedding
import umap.umap_ as umap
```

```python
# Load the dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
# Standardize the features
X_standardized = (X - np.mean(X, axis=0)) / np.std(X, axis=0)
```

```python
# Function to perform dimensionality reduction and create a scatter plot
def reduce_and_plot(method, data, title, labels):
    # Perform dimensionality reduction
    X_reduced = method.fit_transform(data)
    # Create a scatter plot
    sns.scatterplot(x=X_reduced[:, 0], y=X_reduced[:, 1], hue=labels, palette='viridis')
    plt.title(title)
    plt.show()
```

```python
# Apply PCA
reduce_and_plot(PCA(n_components=2), X_standardized, 'PCA reduction of Iris dataset', iris.target)
```

```python
# Apply t-SNE
reduce_and_plot(TSNE(n_components=2, random_state=42), X_standardized, 't-SNE reduction of Iris dataset', iris.target)
```

```python
# Apply MDS
reduce_and_plot(MDS(n_components=2, random_state=42), X_standardized, 'MDS reduction of Iris dataset', iris.target)
```

```python
# Apply Isomap
reduce_and_plot(Isomap(n_components=2), X_standardized, 'Isomap reduction of Iris dataset', iris.target)
```

```python
# Apply LLE
reduce_and_plot(LocallyLinearEmbedding(n_components=2), X_standardized, 'LLE reduction of Iris dataset', iris.target)
```

```python
# Apply UMAP
reduce_and_plot(umap.UMAP(n_components=2, random_state=42), X_standardized, 'UMAP reduction of Iris dataset', iris.target)
```
